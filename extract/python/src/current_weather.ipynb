```python
import requests
import os
import json
import datetime

# 環境変数の読み込み
%load_ext dotenv
%dotenv

# TEMPファイル出力先
TEMP_FILE_DIR = os.environ.get("TEMP_FILE_DIR")

# APIキー
API_KEY = os.environ.get("OWM_API_KEY")
print(API_KEY)
```

```python
# 対象とする都道府県
pref_dict = {
    "東京都":{
        "id":"1850144"
    },
    "神奈川県":{
        "id":"1860291"
    },
}

# 気候情報のJSON
weather_jsons = {}

# 都道府県ごとの気候を取得する
for pref_key,pref_values in pref_dict.items():

    # リクエスト
    response = requests.get(
            "https://api.openweathermap.org/data/2.5/weather",
            params={
                "id": pref_values["id"],
                "appid": API_KEY,
                "units": "metric",
                "lang": "ja",
            },
        )
    # 都道府県名をキーにして、JSONに追加する
    weather_jsons[pref_key] = json.loads(response.text)
```

```python
today = datetime.datetime.today().strftime("%Y%m%d")
filename = f"weather_{today}.json"

with open(os.path.join(TEMP_FILE_DIR,filename),mode="w",encoding="utf-8") as f:
    f.write(json.dumps(weather_jsons,indent = 2,ensure_ascii=False))
```
